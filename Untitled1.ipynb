```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from google.colab import data_table
```

```python
df = pd.read_csv('/content/cars_engage_2022.csv')
```

```python
df = df.loc[df.Boot_Space.notna(),['Make','Model','Variant','Showroom_Price','Fuel_Tank_Capacity','Fuel_Type','Power','Torque','Boot_Space','City_Mileage']]
df["City_Mileage"] = pd.to_numeric(df["City_Mileage"],errors='coerce')
df["Showroom_Price"] = pd.to_numeric(df["Showroom_Price"])
```

```python
df=df.loc[df.Fuel_Tank_Capacity.notna()]
```

```python
df=df.loc[df.City_Mileage.notna()]
```

```python
class_trans = OneHotEncoder(categories='auto',drop=None,sparse=True,handle_unknown='ignore')
```

```python
class_trans.fit_transform(df)
```

```python
class_trans.categories_
```

```python
le= preprocessing.LabelEncoder()
```

```python
le.fit(df.Make)
```
